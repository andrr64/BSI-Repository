```python
import cv2
import numpy as np
from scipy import signal
from matplotlib import pyplot as plt
```

```python
imgc = cv2.imread('kucing.jpeg')
img = cv2.cvtColor(imgc, cv2.COLOR_BGR2GRAY)
```

```python
kernel = np.array(
    [
      [0,-1,0],
      [-1, 5, -1],
      [0,-1,0],
    ],
    dtype='float'
)
```

```python
kernel = np.rot90(kernel, 2)
```

```python
print(img.shape, kernel.shape)
f=signal.convolve2d(img, kernel, 'valid')
print(np.min(f))
```

```python
plt.subplot(121),plt.imshow(img, 'gray'),plt.title('Citra Asli')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(f, 'gray'), plt.title('Citra gaussian blur')
plt.xticks([]), plt.yticks([])
plt.show()
```
