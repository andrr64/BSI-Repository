# Deteksi Tepi Menggunakan cv2
Pada pertemuan kali ini mahasiswa diminta untuk melakukan deteksi tepi pada sebuah gambar menggunakan cv2
<br>

### 1. Import Pustaka
Impor pustaka cv2 (opencv-python) dan matplotlib.pyplot. Pustaka cv2 digunakan untuk ... sedangkan matplotlib.pyplot digunakan untuk menampilkan gambar sebelum dan asesudah

```python
import cv2 #import pustaka opencv-python
import matplotlib.pyplot as plt #import pustaka matplotlib untuk menampilkan gambar
```

### 2. Membuka File Citra
Pada tahap ini kita akan membuka file citra yang ingin di deteksi tepinya. Buat variabel bernama ***citra*** yang nilainya adalah data foto asli, kita bisa menggunakan fungsi _cv2.imread(url, flag)_ untuk mengambil data foto lalu memasukkanya ke variabel ***citra***

```python

citra = cv2.imread("kucing.jpg", 0) # membaca gambar
```

### 3. Deteksi Tepi dengan CV2
Untuk mendeteksi tepi kita bisa menggunakan fungsi *cv2.Canny()*. Caranya buat variabel **tepi** lalu *assign value-nya* dengan *cv2.Canny(citra, 100, 200)*

```python

tepi = cv2.Canny(citra, 100, 200) # mendeteksi tepi menggunakan fungsi cv2.canny dan hasilnya disimpan di variabel tepi
```

```python
# plotting untuk menampilkan citra asli
plt.subplot(121)
plt.imshow(citra, cmap='gray')
plt.title('Citra Asli')
plt. xticks([])
plt. yticks([])

# plotting untuk menampilkan citra hasil deteksi tepi
plt.subplot(122)
plt.imshow(tepi, cmap='gray')
plt.title('Deteksi Tepi')
plt.xticks([])
plt.yticks([])

# tampilkan gambar
plt.show()
```
